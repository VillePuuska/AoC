```python
def print_grid(grid: list[list[str]]):
    print("\n".join("".join(line) for line in grid))


def plot_grid(
    grid: list[list[str]],
    colors: dict[str, int],
    save: bool = False,
    filepath: str = "images/plot.png",
) -> None:
    arr = np.zeros((len(grid), len(grid[0])))
    for r, row in enumerate(grid):
        for c, char in enumerate(row):
            if char in colors:
                arr[r, c] = colors[char]
    plt.xticks([])
    plt.yticks([])
    if save:
        plt.imsave(filepath, arr)
    else:
        plt.imshow(arr)


def plot_objects(
    object_lists: list[list[tuple[int, int]]],
    colors: list[int],
    x_limit: int,
    y_limit: int,
    save: bool = False,
    filepath: str = "images/plot.png",
) -> None:
    arr = np.zeros((y_limit, x_limit))
    for objects, color in zip(object_lists, colors):
        for obj in objects:
            arr[y_limit - 1 - obj[1], obj[0]] = color
    plt.xticks([])
    plt.yticks([])
    if save:
        plt.imsave(filepath, arr)
    else:
        plt.imshow(arr)
```

```python
def get_lines() -> list[str]:
    lines = []
    with open(f"./data/{today}.txt") as f:
        while line := f.readline():
            lines.append(line.rstrip())
    return lines


def get_lines_no_strip() -> list[str]:
    lines = []
    with open(f"./data/{today}.txt") as f:
        while line := f.readline():
            lines.append(line[:-1])
    return lines


def get_grid() -> list[list[str]]:
    grid = []
    with open(f"./data/{today}.txt") as f:
        while line := f.readline():
            grid.append([c for c in line.rstrip()])
    return grid


def parse_nums(s: str) -> list[int]:
    return [int(x) for x in re.findall(r"-?\d+", s)]


def get_nums() -> list[list[int]]:
    lines = get_lines()
    return [parse_nums(line) for line in lines]


def is_inside_grid(coords: tuple[int, int], grid: list[list[Any]]) -> bool:
    return coords[0] in range(len(grid)) and coords[1] in range(len(grid[0]))
```

```python
lines = get_lines()
```

```python
nums: list[list[int]] = []
for line in lines[:-1]:
    nums.append([int(num) for num in line.split()])

operations = lines[-1].split()
```

```python
part_1 = 0

for (problem, operation) in enumerate(operations):
    problem_nums = [nums[row][problem] for row in range(len(nums))]

    match operation:
        case "+":
            part_1 += reduce(lambda x, y: x + y, problem_nums, 0)
        case "*":
            part_1 += reduce(lambda x, y: x * y, problem_nums, 1)

part_1
```

```python
def column_to_num(col: int, lines: list[str]) -> int:
    digits = []
    for row in range(len(lines) - 1):
        c = lines[row][col]
        if c != " ":
            digits.append(c)
    return int("".join(digits))
```

```python
lines = get_lines_no_strip()
```

```python
part_2 = 0

col = len(lines[0]) - 1
nums = []
while col >= 0:
    nums.append(column_to_num(col, lines))
    match lines[-1][col]:
        case "+":
            part_2 += reduce(lambda x, y: x + y, nums, 0)
            col -= 1
            nums = []
        case "*":
            part_2 += reduce(lambda x, y: x * y, nums, 1)
            col -= 1
            nums = []
    col -= 1

part_2
```
